## 1. Tensor
    - numpy의 array와 사용법이 유사하다

```
import torch
import numpy as np
```

```
x = torch.empty(5, 4)
print(x)
```

```
torch.ones(3, 3)
```

```
# 임의의 난수를 가지는 tensor생상
torch.rand(5, 6)
```

```
## 리스트 또는 array를 텐서로 변한
## np.array() 하는것처럼 torch.tensor() 사용하면되


l = [13, 4]
r = np.array([4, 56, 7])

torch.tensor(l)
torch.tensor(r)
```

```

x = torch.rand(2, 2)
y = torch.rand(2, 2)


## 텐서 사이즈알아보가
x.size()
```

```
## 텐서타입
type(x)
```

```
## 텐서 더하는 방법
print(x + y)
print(x.add(y))
print(y.add(x))
```

```
## 더하는 방법인데.. 판다스의 inplace=True처럼 더해주는것
y.add_(x)
print(y)
```

```
## 텐서의 원소에 접근하는 방법
## 판다스와 넘파이와 같이 접근하면된다
print(x[1, 1])
print(x[:, 1])
```

```
x = torch.rand(8, 8)
print(x)
## numpy의 reshape와 같은기능 : view
## Fully Connected Layer로 텐서를 전달할떄, 텐서를 1차원으로 만들어서 넘겨주는데 그떄 사용한다.
print(x.view(64))
print(x.view(-1, 16)) 
```

```
## tensor를 numpy로 변환하는 방법
y = x.numpy()
print(y)
print(type(y))
```

```
## 텐서의 원소를 텐서가아닌 값(Value : 스칼라)로 출력할떄
print(x[1, 1].item())
print(x[1, 1])
```
