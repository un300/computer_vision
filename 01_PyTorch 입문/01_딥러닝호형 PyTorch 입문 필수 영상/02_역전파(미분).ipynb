### 딥러닝 구조의 흐름

#### (1) 변수 선연 (데이터입력)

(requires_grad = True)
아래를 보면 텐서하나를 만들어 `x`에 할당했다. 그 중 (requires_grad = True)라는 옵션은 `x`에 관한 연산들을 모두 추적할 수 있도록 해주는 옵션이다.
연산을 추적하여 미분을 진행한다.

```
x = torch.ones(2, 2, requires_grad=True)
print(x)
```

#### (2) 모델내 연산 예측값 산출

```
y = x + 1
print(y)
```

#### (3) 손실함수 계산 및 손실 산출

```
z = 2*y**2
res = z.mean()
print(res)
```

```
## 위에서 구한 연산을 기점으로 미분을 하는데 역전파를 하겟다
res.backward()
```

```
print(x.grad)
```
